```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
## vocabulary size
voc_size = 5000
## we may mention 10,000 as well, but if voc_size will increase
## onehot encoding representation quite big for this vocabulary
```

```python
# one hot representation
messages=X.copy()
```

```python
messages[0:10]
```

```python
messages.reset_index(inplace=True,drop=True)
```

```python
messages[0:10]
```

```python
messages['title'][7]
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
## Dataset Preprocessing
from nltk.stem.porter import PorterStemmer ##stemming purpose
ps = PorterStemmer()
corpus = []
for i in range(0,len(messages)):
  review = re.sub('[^a-zA-Z]',' ', messages['title'][i])
  review = review.lower()
  review = review.split()

  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
corpus[1]
```

```python
onehot_repr = [one_hot(words,voc_size) for words in corpus]
onehot_repr
```

```python
corpus[1]
```

```python
onehot_repr[1]
```

```python
## Embedding representation
```

```python
sent_length =30
embedded_docs = pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
embedded_docs[0]
```

```python
embedded_docs[1]
```

```python
# for tf-idf --> word embedding or word 2vec is not completly sufficent
# for LSTM --> this word embedding is sufficient
```

```python
# creating model
from tensorflow.keras.layers import Dropout
embedding_vector_features = 50 ## 40,50,100 like hyper parameter ## 50 feature representation
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3)) ## it is using for reducing the over fitting
model.add(LSTM(100)) ## 100, 200 it is our choice hyperparameters
model.add(Dense(1,activation='sigmoid'))
model.compile(loss = 'binary_crossentropy',optimizer ='adam',metrics=['accuracy']) ## fake or not need to predict this is binary classification model that's
# used binary cross entory
print(model.summary())
```

```python
import numpy as np
X_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_final,y_final,test_size=0.33,random_state=42)
```

```python
#Model training
```

```python
model.fit(X_train,y_train,validation_data = (X_test,y_test),epochs=10,batch_size=64)
```

```python
### Performence MEtrics and Accuracy
y_pred = model.predict(X_test)
```

```python
y_pred =np.where(y_pred > 0.5, 1,0) ###AUC ROC Curve
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
